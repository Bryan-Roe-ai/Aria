# Aria Contributing & Development Workflows

**Complete guide for contributing to Aria, development workflows, collaboration patterns, and best practices.**

Learn how to implement features, collaborate with the team, and maintain code quality.


## Development Workflow Overview

### Feature Development Cycle

```
1. Planning          → Feature spec, dependencies, design
2. Implementation   → Code changes, tests, documentation
3. Local Testing    → Unit tests, integration tests, manual testing
4. Code Review      → PR review, feedback, revisions
5. Merge            → Merge to main, CI/CD runs
6. Deploy           → Auto-deploy to staging, then production
7. Monitoring       → Watch metrics, logs, alerts
8. Done             → Feature live, document lessons learned
```

## Setting Up Your Development Environment

### Initial Setup

```bash
# Clone repository
git clone https://github.com/Bryan-Roe/Aria.git
cd Aria

# Create feature branch
git checkout -b feature/my-feature-name

# Create virtual environment
python3.12 -m venv .venv
source .venv/bin/activate

# Install dev dependencies
pip install -r requirements-dev.txt

# Verify setup
python scripts/fast_validate.py

# You're ready!
echo "Development environment ready ✓"
```

```json
// .vscode/settings.json (create if missing)
{
    "python.defaultInterpreterPath": "${workspaceFolder}/.venv/bin/python",
    "python.linting.enabled": true,
    "python.linting.ruffEnabled": true,
    "python.formatting.provider": "black",
    "[python]": {
        "editor.formatOnSave": true,
        "editor.codeActionsOnSave": {
            "source.organizeImports": true
        }
    }
}
```

## Pull Request Workflow

### Creating a Pull Request

**PR Checklist:**

```markdown
## Checklist

- [ ] Tests pass: `python scripts/test_runner.py --unit`
- [ ] Lint passes: `make lint`
- [ ] Type check passes: `make type-check`
- [ ] Documentation updated
- [ ] No hardcoded secrets
- [ ] Commits are logical and well-formed
- [ ] Ready for review

## Description

### Problem

What problem does this solve?

### Solution

How does it solve it?

### Testing

How was this tested?

### Breaking Changes

Any breaking changes?

### Related Issues

Fixes #123
Relates to #456
```

**Push and Create PR:**

```bash
# Make sure you're on your feature branch
git branch
# * feature/my-feature-name
#   main

# Push to GitHub
git push origin feature/my-feature-name

# Create PR via GitHub CLI
gh pr create --title "Feature: My Feature" --body "Detailed description" --base main

# Or use GitHub web interface
# https://github.com/Bryan-Roe/Aria/compare/main...your-branch
```

### During Code Review

**Responding to Feedback:**

```bash
# 1. Update code based on review comments
# Edit files locally

# 2. Run tests to ensure changes are good
python scripts/test_runner.py --all

# 3. Commit changes (can add to same commit or new one)
git add .
git commit -m "Address review feedback: explain what changed"

# 4. Push updates
git push origin feature/my-feature-name
# GitHub automatically updates PR

# 5. Respond to comments
# Use GitHub PR comments to explain changes
```

**When Changes Are Requested:**

```
Don't be discouraged! Code review is collaborative.
- Reviewers are helping improve the code
- They have domain expertise
- Their feedback makes the feature better

Approach:
1. Understand the feedback
2. Ask questions if unclear
3. Make changes
4. Explain your changes
5. Move forward together
```

### Merging the PR

```bash
# Ensure all checks pass:
# ✓ Tests pass
# ✓ Code reviewed and approved
# ✓ No conflicts with main
# ✓ Documentation updated

# Merge via GitHub (web or CLI)
gh pr merge --squash  # Squash commits
# or
gh pr merge           # Keep commit history

# After merge, delete branch
git branch -d feature/my-feature-name
git push origin -d feature/my-feature-name
```

## Collaboration Best Practices

### Communication

**Clear Commit Messages:**

```
Good:
  "Add semantic memory retrieval for chat context"
  "Fix null pointer in Aria gesture handler"
  "Update deployment docs for AWS"

Bad:
  "Fixed stuff"
  "Work in progress"
  "asdf"
```

**Good PR Descriptions:**

```markdown
## What

Adds real-time chat memory to improve context retention

## Why

Users were losing context after 5-10 messages

## How

- Added semantic memory module in shared/
- Injected memory into chat provider
- Added tests for memory retrieval
- Updated API docs

## Testing

- Unit tests: 15 new tests, all passing
- Integration tests: Tested with all providers
- Manual testing: Verified with 100+ message conversations
```

**Issues & Discussions:**

```bash
# Before implementing something big, open an issue
gh issue create --title "Proposal: Feature X" --body "Description and rationale"

# Discuss architecture before coding
# Get feedback early
# Avoid big surprising PRs

# Reference issues in commits
git commit -m "Implement feature X (#123)"
```

## Advanced Workflows

### Workflow 1: Fixing a Bug

```bash
# 1. Create branch from main
git checkout main
git pull
git checkout -b fix/bug-name

# 2. Write failing test that reproduces bug
pytest tests/test_bug_reproduction.py
# Should fail ✗

# 3. Fix the bug
# Edit file to fix issue

# 4. Verify test now passes
pytest tests/test_bug_reproduction.py
# Should pass ✓

# 5. Run full test suite
python scripts/test_runner.py --all

# 6. Commit and create PR
git add .
git commit -m "Fix: Describe bug fix (fixes #123)"
git push origin fix/bug-name

# 7. Create PR linking to issue
gh pr create --title "Fix: Bug name" --body "Fixes #123"
```

### Workflow 2: Performance Optimization

```bash
# 1. Measure baseline performance
python scripts/benchmark.py --baseline
# Save baseline.json

# 2. Implement optimization
# Edit performance-critical code

# 3. Measure new performance
python scripts/benchmark.py
# Compare with baseline

# 4. Document improvement
# Add comment showing: before X ms → after Y ms

# 5. Verify tests still pass
python scripts/test_runner.py --unit --coverage

# 6. Commit with metrics
git commit -m "Optimize: Chat loading time 500ms → 50ms"
```

### Workflow 3: Refactoring

```bash
# 1. Create refactoring branch
git checkout -b refactor/module-name

# 2. Refactor code
# Change implementation, not behavior

# 3. Run tests (should all pass)
python scripts/test_runner.py --unit
# No new tests needed, just restructuring

# 4. Run linter to catch style issues
make lint
make format

# 5. Commit with explanation
git commit -m "Refactor: Extract X into separate module for clarity"

# 6. PR with rationale
# "This refactoring improves code organization by..."
```
