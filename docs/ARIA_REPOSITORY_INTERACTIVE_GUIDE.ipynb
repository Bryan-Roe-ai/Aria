## Section 4: Quick Reference Cards

### Core Modules at a Glance

#### 🎭 Interactive Character (Aria)
```
Location: apps/aria/
Port: 8080
Entry Point: python server.py

Key Files:
├── server.py (HTTP API)
├── aria_controller.js (UI logic)
├── AUTO-EXECUTE.md (action contract)
└── index.html (web interface)

Main Endpoints:
GET  /api/aria/state       → Character state (position, objects)
POST /api/aria/command     → Natural language → actions
POST /api/aria/execute     → Run action sequence
GET  /api/aria/schema      → Valid actions definition

Example Command:
"move left and wave"
↓ (processed)
[
  {action: "move", x: 30, duration: 1},
  {action: "gesture", name: "wave", duration: 2}
]
↓ (executed)
Aria moves to position 30, then waves
```

#### 💬 Multi-Provider Chat
```
Location: ai-projects/chat-cli/, shared/chat_providers.py
Entry Point: chat_cli.py --provider <name>

Providers (in priority order):
1. Explicit selection (CLI flag)
2. LM Studio (localhost:1234)
3. Ollama (localhost:11434)
4. Azure OpenAI (cloud)
5. OpenAI (cloud)
6. Local Echo (always available)

Key Functions:
detect_provider()           → Auto-detect available provider
ChatProvider.complete()     → Sync completion
ChatProvider.stream()       → SSE streaming

Config:
DEFAULT_AI_PROVIDER         → Default provider name
LMSTUDIO_BASE_URL           → LM Studio location
OLLAMA_BASE_URL             → Ollama location
AZURE_OPENAI_API_KEY        → Azure credentials
```

#### 🧠 Semantic Memory
```
Location: shared/chat_memory.py
Purpose: RAG (Retrieval-Augmented Generation)

Workflow:
1. Embed message → multi-qa-MiniLM-L6-cos-v1
2. Store in DB (SQL or Cosmos)
3. Search for similar messages
4. Inject top-K into prompt
5. Send to LLM
6. Store response

Key Functions:
generate_embedding(text)    → Create vector
fetch_similar(query, k=5)   → Find related messages
store_memory(session_id, ...) → Persist to DB

Performance:
- Embedding: ~10ms per message
- Search: ~20ms for top-5 results
- Total: ~50ms per chat turn (if cached)
```

#### ⚛️ Quantum Computing
```
Location: ai-projects/quantum-ml/
Port: MCP (Model Context Protocol)

Key Components:
├── quantum_sampler.py    → Circuit generation
├── quantum_embeddings.py → Quantum embedding transform
├── quantum_router.py     → Backend selection
└── config.py             → Cost/safety configuration

Backends (in order):
1. Qiskit Aer (free, local, fast)
2. Azure Quantum Simulator (free, cloud)
3. ionQ QPU (paid, real quantum hardware)

Cost Control:
azure_confirm_cost: true   → Require confirmation for real QPU
simulator_only: true       → Never use paid backends
max_shots: 1000           → Limit measurement samples

Entry Point:
quantum_mcp_server.py      → MCP server with safety gates
```

#### 🎓 Training & LoRA
```
Location: ai-projects/lora-training/microsoft_phi-silica-3.6_v1/
Orchestrator: scripts/autonomous_training_orchestrator.py

Workflow:
1. Discover datasets in datasets/
2. Select epochs (adaptive based on history)
3. Load base model (Phi-3.5)
4. Attach LoRA adapter
5. Train with PEFT
6. Evaluate metrics
7. Auto-promote if accuracy > 0.90

Configuration:
config/autonomous_training.yaml
├── cycle_interval_minutes: 30
├── epochs_progression: [25, 50, 100, 200]
├── min_accuracy_threshold: 0.70
├── auto_promote: true
└── dataset_minimum: 5

State:
data_out/autonomous_training_status.json
├── cycles_completed
├── best_accuracy
├── performance_history[]
└── dataset_inventory

Run Command:
python scripts/autonomous_training_orchestrator.py
```

#### 🗄️ Database Layer
```
Location: shared/sql_engine.py, shared/cosmos_client.py

SQL Configuration:
QAI_DB_CONN="sqlite:///local.db"    (dev, local)
QAI_DB_CONN="postgresql://..."      (production)
QAI_DB_CONN="mssql://..."           (Azure SQL)
QAI_SQL_POOL_SIZE=10                (connection pool size)

Cosmos Configuration:
QAI_ENABLE_COSMOS=true              (enable if needed)
COSMOS_ENDPOINT="https://..."       (endpoint URL)
COSMOS_KEY="..."                    (primary key)
COSMOS_DATABASE="aria_db"           (database name)

Health Check:
curl http://localhost:7071/api/ai/status | jq .database
Shows: connection_available, pool_saturation, response_time_ms
```

## Section 5: Common Workflows

### Workflow 1: Understanding a Chat Request

**Question:** "How does a chat message flow through the system?"

**Path:**
1. User sends message to `/api/chat` (function_app.py)
2. Function detects provider (shared/chat_providers.py)
3. Message embedded for memory (shared/chat_memory.py)
4. Similar past messages retrieved (vector search)
5. Prompt augmented with context (RAG)
6. Sent to provider (Azure OpenAI, OpenAI, local, etc.)
7. Response streamed back via SSE
8. Response stored in memory
9. Logged to database (shared/db_logging.py)

**Files to understand:**
- `function_app.py` → `/api/chat` endpoint
- `shared/chat_providers.py` → Provider selection
- `shared/chat_memory.py` → Memory injection
- `shared/token_utils.py` → Context pruning
- `ai-projects/chat-cli/src/chat_cli.py` → CLI equivalent

**Time complexity:**
- Provider detection: 10-100ms (probes if needed)
- Message embedding: 10ms
- Vector search: 20ms
- LLM call: 1-5 seconds
- Total per message: 1-5.5 seconds

**Diagram:**
See ARCHITECTURE_DIAGRAMS.md → "Chat Message Processing"

```python
class ClaudeProvider(ChatProvider):
    def __init__(self, api_key: str):
        self.api_key = api_key
        # Initialize client
    
    async def complete(self, prompt: str) -> str:
        # Call Claude API
        pass
    
    async def stream(self, prompt: str):
        # Stream responses
        async for token in stream_response:
            yield token
```

2. **Add to provider registry** in `detect_provider()`:
```python
def detect_provider():
    # ... existing checks ...
    if "CLAUDE_API_KEY" in env:
        return ClaudeProvider(env["CLAUDE_API_KEY"])
```

### Workflow 3: Deploying a Trained Model

**Question:** "How do I promote a newly trained LoRA model to production?"

**Automatic (Recommended):**
```yaml
# config/autonomous_training.yaml
auto_promote: true              # Auto-deploy when accuracy > threshold
min_accuracy_for_promotion: 0.90  # Must beat this score
```

Then the system automatically:
1. Trains model in cycle
2. Evaluates accuracy
3. If accuracy > 0.90:
   - Saves adapter to `deployed_models/current/`
   - Updates `lora_infer_bridge.py` to load new adapter
   - Tests with `/api/quantum-llm/chat`
   - Logs promotion event

**Manual Deployment:**
1. **Check status**:
```bash
curl http://localhost:7071/api/ai/status | jq .training
# Shows: latest_model, accuracy, promoted_at
```

2. **Verify model**:
```bash
# List trained adapters
ls -la ai-projects/lora-training/microsoft_phi-silica-3.6_v1/outputs/

# Check metrics
cat data_out/autonomous_training_status.json | jq .best_accuracy
```

3. **Promote manually**:
```bash
# Copy adapter to deployed folder
cp ai-projects/lora-training/.../adapter_model.safetensors deployed_models/current/
cp ai-projects/lora-training/.../adapter_config.json deployed_models/current/
```

4. **Test in production**:
```bash
# Chat with new model
curl -X POST http://localhost:7071/api/chat \
  -H "Content-Type: application/json" \
  -d '{"message": "Hello with new model", "provider": "lora"}'
```

5. **Monitor rollout**:
- Check dashboard at `http://localhost:3000`
- View metrics: `python scripts/training_analytics.py`
- Watch logs: `tail -f data_out/autonomous_training.log`

**Rollback if needed:**
```bash
# Restore previous model
cp deployed_models/backups/previous/* deployed_models/current/
systemctl restart aria-api  # Or redeploy
```

**Related files:**
- `scripts/autonomous_training_orchestrator.py` (promotion logic)
- `lora_infer_bridge.py` (adapter loading)
- `config/autonomous_training.yaml` (settings)
- `deployed_models/` (current/backups folders)

```bash
# List all test markers
pytest --markers

# Run only fast tests (skip slow/azure/integration)
pytest -m "not slow and not azure" tests/

# Run only unit tests
pytest -m "unit" tests/
```

### Problem: "Provider not detected, falling back to local"

**Diagnosis:**
```bash
# Check what's actually running
netstat -tuln | grep -E "1234|11434"  # LM Studio / Ollama
echo $LMSTUDIO_BASE_URL               # Check env var
echo $AZURE_OPENAI_API_KEY            # Check Azure credentials
```

**Solution:**
```bash
# Option 1: Start LM Studio locally
# Download from lmstudio.ai and run

# Option 2: Start Ollama
ollama serve

# Option 3: Set Azure credentials
export AZURE_OPENAI_API_KEY="your_key"
export AZURE_OPENAI_ENDPOINT="https://..."
export AZURE_OPENAI_DEPLOYMENT="your_deployment"
export AZURE_OPENAI_API_VERSION="2024-02-15-preview"

# Option 4: Use local echo (always works, but zero functionality)
# Just works as-is, returns echo responses
```

**See:** REPOSITORY_ARCHITECTURE.md → "Provider Detection Flow"

### Problem: "Database connection pool saturated (80%+)"

**Diagnosis:**
```bash
# Check pool status
curl http://localhost:7071/api/ai/status | jq .database.saturation

# See active connections
# For SQLite: check file locks
# For PostgreSQL: SELECT count(*) FROM pg_stat_activity;
```

**Solution:**
```bash
# Increase pool size
export QAI_SQL_POOL_SIZE=20  # Default is 10

# Or switch to Cosmos DB for horizontal scaling
export QAI_ENABLE_COSMOS=true
export COSMOS_ENDPOINT="https://..."
export COSMOS_KEY="..."

# Restart API
func host start --port 7071
```

**See:** REPOSITORY_TOPICS.md → "Integration & Infrastructure"

### Problem: "Quantum job timeout / costs too much"

**Diagnosis:**
```bash
# Check config
cat config/quantum_autorun.yaml | grep -E "simulator|cost|timeout"

# See what's configured
curl http://localhost:7071/api/quantum-llm/status | jq
```

**Solution:**
```yaml
# config/quantum_autorun.yaml
simulator_only: true                    # Never use real QPU
azure_confirm_cost: false               # Reject expensive jobs
max_shots: 100                          # Limit measurements
circuit_depth_limit: 20                 # Keep circuits simple
```

**Then restart:**
```bash
python scripts/quantum_autorun.py --dry-run  # Validate
python scripts/quantum_autorun.py            # Run
```

**See:** ARCHITECTURE_DIAGRAMS.md → "Quantum Job Execution"
