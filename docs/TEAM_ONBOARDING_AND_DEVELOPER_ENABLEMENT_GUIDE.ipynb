````python
```xml
<VSCode.Cell language="markdown">
# 🎓 Team Onboarding & Developer Enablement Guide

**Building effective onboarding programs and developer experience**

This guide covers:
- Onboarding roadmap and timeline
- Knowledge documentation
- Environment setup automation
- Pair programming and mentoring
- Metrics and success criteria
- Technical skill assessment
- Developer productivity tracking
</VSCode.Cell>

<VSCode.Cell language="markdown">
## 1. Onboarding Timeline

### Week 1: Foundation (Days 1-5)
```
Day 1: Welcome & Setup
├─ Meet team
├─ Set up workstation
├─ Access to repos, docs, Slack
├─ Local environment setup
└─ First commit (docs update)

Day 2-3: Architecture Overview
├─ System design walkthrough
├─ Code repository tour
├─ CI/CD pipeline explanation
├─ Database overview
└─ API contract review

Day 4-5: First Contribution
├─ Pick simple issue (good-first-issue)
├─ Pair program with mentor
├─ Submit PR with review
├─ Learn code review process
└─ Merge first PR
```

### Week 2-4: Deep Dives (Days 6-20)
```
Week 2: Development Workflow
├─ Local testing and debugging
├─ Git workflows (branching, merging)
├─ Code review standards
├─ Deployment process
└─ Environment configuration

Week 3: Technical Systems
├─ Database architecture
├─ Caching layer
├─ Messaging/queues
├─ External integrations
└─ Monitoring & observability

Week 4: Onboarding Completion
├─ Ownership of first small feature
├─ Independent code review participation
├─ On-call shadow session
├─ Knowledge assessment
└─ Feedback & retrospective
```

### Onboarding Checklist
```python
import json
from datetime import datetime, timedelta

onboarding_checklist = {
    'name': 'New Engineer Onboarding',
    'created': datetime.now().isoformat(),
    'tasks': [
        {
            'phase': 'Day 1',
            'items': [
                {'task': 'Account setup', 'due': 1, 'owner': 'HR'},
                {'task': 'Laptop setup', 'due': 1, 'owner': 'IT'},
                {'task': 'Office tour', 'due': 1, 'owner': 'Team'},
                {'task': 'Repo access', 'due': 1, 'owner': 'Security'},
            ]
        },
        {
            'phase': 'Week 1',
            'items': [
                {'task': 'Local env setup', 'due': 5, 'owner': 'Mentor'},
                {'task': 'First PR', 'due': 5, 'owner': 'Mentor'},
                {'task': 'Codebase overview', 'due': 3, 'owner': 'Tech Lead'},
            ]
        },
        {
            'phase': 'Week 4',
            'items': [
                {'task': 'On-call shadow', 'due': 20, 'owner': 'SRE'},
                {'task': 'Knowledge assessment', 'due': 20, 'owner': 'Manager'},
                {'task': 'Onboarding complete', 'due': 20, 'owner': 'HR'},
            ]
        }
    ]
}
```
</VSCode.Cell>

<VSCode.Cell language="markdown">
## 2. Environment Setup Automation

### Quick Start Script
```bash
#!/bin/bash
# setup-dev-env.sh - One command to set up everything

set -e

echo "🚀 Setting up development environment..."

# 1. Check prerequisites
command -v git >/dev/null || { echo "Git not installed"; exit 1; }
command -v python3 >/dev/null || { echo "Python3 not installed"; exit 1; }

# 2. Clone repository
if [ ! -d "aria-repo" ]; then
    git clone --depth 1 https://github.com/Bryan-Roe/Aria.git
    cd aria-repo
else
    cd aria-repo
    git pull origin main
fi

# 3. Set up Python environment
if [ ! -d ".venv" ]; then
    python3 -m venv .venv
fi

source .venv/bin/activate

# 4. Install dependencies
pip install --upgrade pip setuptools wheel
pip install -r requirements-dev.txt

# 5. Set up pre-commit hooks
pre-commit install

# 6. Configure local environment
cp local.settings.json.example local.settings.json

# 7. Create databases (if needed)
if [ ! -f "data_out/local.db" ]; then
    python scripts/init_db.py
fi

# 8. Run tests
python -m pytest tests/ -q

echo "✅ Development environment ready!"
echo ""
echo "Quick start:"
echo "  source .venv/bin/activate"
echo "  make start                    # Start all services"
echo "  make test                     # Run tests"
echo "  make lint                     # Check code quality"
```

### Docker Setup (Alternative)
```dockerfile
# Dockerfile.dev - Development environment
FROM python:3.11-slim

WORKDIR /workspace

# Install system dependencies
RUN apt-get update && apt-get install -y \
    git \
    curl \
    postgresql-client \
    && rm -rf /var/lib/apt/lists/*

# Copy dependencies
COPY requirements-dev.txt .
RUN pip install -r requirements-dev.txt

# Copy code
COPY . .

# Set up pre-commit
RUN git init && pre-commit install --install-hooks || true

# Default command
CMD ["/bin/bash"]
```

### Docker Compose
```yaml
# docker-compose.dev.yml
version: '3.9'

services:
  dev:
    build:
      context: .
      dockerfile: Dockerfile.dev
    volumes:
      - .:/workspace
      - /workspace/.venv  # Preserve venv
    environment:
      - PYTHONUNBUFFERED=1
      - DATABASE_URL=postgresql://user:pass@postgres:5432/aria
    depends_on:
      - postgres
      - redis
    ports:
      - "8000:8000"  # API
      - "8080:8080"  # Aria UI

  postgres:
    image: postgres:15
    environment:
      - POSTGRES_DB=aria
      - POSTGRES_USER=user
      - POSTGRES_PASSWORD=pass
    volumes:
      - postgres_data:/var/lib/postgresql/data

  redis:
    image: redis:7

volumes:
  postgres_data:
```

Start everything with: `docker-compose -f docker-compose.dev.yml up -d`
</VSCode.Cell>

<MySCode.Cell language="markdown">
## 3. Knowledge Management

### Documentation Structure
```
docs/
├─ GETTING_STARTED.md           # First read
├─ ARCHITECTURE_OVERVIEW.md     # System design
├─ API_REFERENCE.md             # API docs
├─ DATABASE_SCHEMA.md           # Data model
├─ DEPLOYMENT_GUIDE.md          # How to deploy
├─ TROUBLESHOOTING.md           # Common issues
├─ RUNBOOKS/
│  ├─ how-to-debug-payment-flow.md
│  ├─ how-to-scale-database.md
│  └─ how-to-respond-to-incidents.md
└─ FAQ.md                        # Frequent questions
```

### Knowledge Retention Assessment
```python
class OnboardingAssessment:
    def __init__(self, engineer_name: str):
        self.name = engineer_name
        self.week = 1
    
    def technical_knowledge_test(self):
        """Test technical understanding"""
        
        questions = [
            {
                'question': 'How does the payment API handle retries?',
                'correct_answer': 'Idempotent keys',
                'options': ['Idempotent keys', 'Retry count', 'Time-based', 'None'],
                'importance': 'HIGH'
            },
            {
                'question': 'What monitoring tool tracks production metrics?',
                'correct_answer': 'Datadog',
                'options': ['Datadog', 'New Relic', 'None', 'Grafana'],
                'importance': 'MEDIUM'
            }
        ]
        
        score = 0
        for q in questions:
            # Present question, get answer
            answer = input(f"{q['question']}: ")
            if answer == q['correct_answer']:
                score += 1
        
        total = len(questions)
        percentage = (score / total) * 100
        
        status = "PASS" if percentage >= 80 else "REVIEW"
        
        return {
            'score': score,
            'total': total,
            'percentage': percentage,
            'status': status
        }
    
    def soft_skills_evaluation(self):
        """Team lead evaluates soft skills"""
        
        rubric = {
            'communication': {
                'not_ready': 'Struggles to explain ideas',
                'developing': 'Can explain with help',
                'ready': 'Clear, articulate communicator'
            },
            'collaboration': {
                'not_ready': 'Works in isolation',
                'developing': 'Collaborates with support',
                'ready': 'Proactive team contributor'
            },
            'initiative': {
                'not_ready': 'Needs constant direction',
                'developing': 'Can work on defined tasks',
                'ready': 'Identifies and tackles problems independently'
            }
        }
        
        return rubric
```
</MySCode.Cell>

<MySCode.Cell language="markdown">
## 4. Mentoring & Pair Programming

### Pair Programming Structure
```python
class PairProgrammingSession:
    def __init__(self, driver: str, navigator: str, duration_minutes: int = 90):
        self.driver = driver  # Types code
        self.navigator = navigator  # Reviews, suggests
        self.duration = duration_minutes
    
    def schedule_session(self, task: str, learning_objectives: list):
        """Plan pair programming session"""
        
        return {
            'task': task,
            'driver': self.driver,
            'navigator': self.navigator,
            'duration': self.duration,
            'objectives': learning_objectives,
            'success_criteria': [
                'Task completed',
                'Code reviewed',
                'PR submitted',
                'Learnings documented'
            ]
        }
    
    def run_session(self):
        """Execute with best practices"""
        
        # 1. Kick-off (5 min)
        print("📋 Review objectives and task")
        
        # 2. Setup (5 min)
        print("⚙️ Set up environment together")
        
        # 3. Pair (70 min)
        print("👥 Switch driver/navigator every 20 min")
        print("  - Driver types and explains")
        print("  - Navigator reviews and suggests")
        print("  - Ask questions, discuss trade-offs")
        
        # 4. Wrap-up (10 min)
        print("✅ Review what was learned")
        print("📝 Document takeaways")
    
    def feedback(self, feedback_for: str, observations: list):
        """Collect feedback from session"""
        
        return {
            'feedback_for': feedback_for,
            'date': datetime.now().isoformat(),
            'observations': observations,
            'strengths': [],
            'growth_areas': [],
            'next_session_focus': []
        }
```

### Mentoring Plan
```
Week 1-2: Basics
├─ Local environment
├─ Code structure
└─ First PR

Week 3-4: Independent Tasks
├─ Own small features
├─ Code reviews
└─ Debugging

Week 5-8: Advanced Topics
├─ System design
├─ Performance optimization
├─ Deployment procedures
└─ On-call readiness
```
</MySCode.Cell>

<MySCode.Cell language="markdown">
## 5. Success Metrics & Retention

### Onboarding Success Metrics
```python
def calculate_onboarding_health(engineer_data):
    """Measure onboarding effectiveness"""
    
    metrics = {
        'time_to_first_pr': engineer_data['first_pr_days'],
        'time_to_first_merge': engineer_data['first_merge_days'],
        'pr_review_turnaround': engineer_data['avg_review_time_hours'],
        'independent_task_completion': engineer_data['tasks_completed_alone'],
        'knowledge_test_score': engineer_data['assessment_score'],
        'team_feedback_score': engineer_data['team_feedback'],
        'retention_likelihood': engineer_data['satisfaction_score']
    }
    
    # Healthy onboarding benchmarks:
    healthy = {
        'time_to_first_pr': 7,  # Within first week
        'time_to_first_merge': 14,  # Within 2 weeks
        'pr_review_turnaround': 24,  # Hours
        'independent_task_completion': 5,  # By week 4
        'knowledge_test_score': 80,  # Percentage
        'team_feedback_score': 4,  # Out of 5
        'retention_likelihood': 4.5  # Out of 5
    }
    
    health_score = 0
    for metric, value in metrics.items():
        if value >= healthy[metric]:
            health_score += 1
    
    status = "✅ HEALTHY" if health_score >= 6 else "⚠️ NEEDS SUPPORT"
    
    return {
        'metrics': metrics,
        'health_score': f"{health_score}/7",
        'status': status,
        'recommendations': get_recommendations(metrics, healthy)
    }

def get_recommendations(metrics, healthy):
    """Suggest improvements"""
    
    recommendations = []
    
    if metrics['time_to_first_pr'] > healthy['time_to_first_pr']:
        recommendations.append("Schedule more pair programming sessions")
    
    if metrics['knowledge_test_score'] < healthy['knowledge_test_score']:
        recommendations.append("Additional mentoring on system architecture")
    
    if metrics['team_feedback_score'] < healthy['team_feedback_score']:
        recommendations.append("Facilitate team connection activities")
    
    return recommendations
```

### 30-60-90 Day Plan Template
```
30 Days: Foundation
✅ Environment setup complete
✅ Understand architecture
✅ 3+ PRs merged
✅ Can run tests independently
✅ Familiar with deployment

60 Days: Competency
✅ Own small features end-to-end
✅ Effective code reviews
✅ Debug production issues
✅ Participate in design reviews
✅ Shadow on-call

90 Days: Independence
✅ Pick up complex features
✅ Mentor new engineer
✅ Full on-call capability
✅ Contribute to roadmap
✅ Feedback on process
```
</MySCode.Cell>
```
````
