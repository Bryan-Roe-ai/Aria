````python
```xml
<VSCode.Cell language="markdown">
# 🌐 Edge Computing & IoT Guide

**Building edge-native systems with device management, offline-first sync, and edge ML inference**

This guide covers:
- Edge computing architectures (fog, edge cloud)
- IoT device management and provisioning
- Edge ML inference and on-device models
- Offline-first data synchronization
- Edge-cloud coordination
- IoT protocols (MQTT, CoAP)
- Device telemetry and monitoring
- Security for edge devices
</VSCode.Cell>

<VSCode.Cell language="markdown">
## 1. Edge Architecture Patterns

### Edge-Cloud Coordination
```python
from dataclasses import dataclass
from typing import Dict, List
from datetime import datetime

@dataclass
class EdgeNode:
    """Represents an edge computing node"""
    node_id: str
    location: str
    capabilities: List[str]  # e.g., ["gpu", "high-memory"]
    available_memory: int
    available_storage: int
    connected: bool = True
    last_heartbeat: datetime = None

class EdgeCloudCoordinator:
    """Coordinate between edge nodes and cloud"""
    
    def __init__(self):
        self.edge_nodes: Dict[str, EdgeNode] = {}
        self.workload_assignments = {}
    
    def register_edge_node(self, node: EdgeNode):
        """Register new edge node"""
        self.edge_nodes[node.node_id] = node
        print(f"Registered edge node: {node.node_id}")
    
    def assign_workload(self, workload_id: str, workload_type: str, required_capabilities: List[str]):
        """Assign workload to suitable edge node"""
        # Find suitable edge node
        for node_id, node in self.edge_nodes.items():
            if not node.connected:
                continue
            
            # Check capabilities match
            if all(cap in node.capabilities for cap in required_capabilities):
                # Check resources available
                if node.available_memory > 1000 and node.available_storage > 5000:
                    self.workload_assignments[workload_id] = node_id
                    print(f"Assigned {workload_id} to edge node {node_id}")
                    return True
        
        # If no suitable edge node, send to cloud
        print(f"No suitable edge node found. Sending {workload_id} to cloud")
        return False
    
    def check_node_health(self):
        """Check health of all edge nodes"""
        import time
        
        current_time = datetime.now()
        
        for node_id, node in self.edge_nodes.items():
            if node.last_heartbeat:
                elapsed = (current_time - node.last_heartbeat).total_seconds()
                
                if elapsed > 30:  # 30 second timeout
                    node.connected = False
                    print(f"Edge node {node_id} disconnected (no heartbeat)")
    
    def get_latency_estimate(self, node_id: str) -> int:
        """Get estimated latency to edge node"""
        # Latency depends on network distance
        # Typically 1-50ms for edge, 100+ ms for cloud
        return 10  # Simplified
```

### Hierarchical Edge Architecture
```python
class HierarchicalEdgeArchitecture:
    """Multi-tier edge computing hierarchy"""
    
    def __init__(self):
        self.cloud_layer = "cloud"
        self.regional_edge = {}  # Regional data centers
        self.local_edge = {}      # Local edge nodes
        self.devices = {}          # IoT devices
    
    def route_data(self, data: Dict, source_device: str) -> str:
        """Route data through hierarchy"""
        # Determine where to process data
        
        if self._needs_cloud_processing(data):
            return self.cloud_layer
        elif self._needs_regional_processing(data):
            return self._get_nearest_regional_edge(source_device)
        else:
            return self._get_nearest_local_edge(source_device)
    
    def _needs_cloud_processing(self, data: Dict) -> bool:
        """Check if data needs cloud processing"""
        # Check data size, complexity, etc.
        return data.get('complexity', 'low') == 'high'
    
    def _needs_regional_processing(self, data: Dict) -> bool:
        """Check if data needs regional processing"""
        return data.get('complexity', 'low') == 'medium'
    
    def _get_nearest_regional_edge(self, source_device: str) -> str:
        """Get nearest regional edge node"""
        return "regional-edge-1"  # Simplified
    
    def _get_nearest_local_edge(self, source_device: str) -> str:
        """Get nearest local edge node"""
        return "local-edge-1"  # Simplified
```
</VSCode.Cell>

<VSCode.Cell language="markdown">
## 2. IoT Device Management

### Device Provisioning
```python
from uuid import uuid4
import json

class DeviceProvisioner:
    """Provision and manage IoT devices"""
    
    def __init__(self):
        self.devices = {}
        self.certificates = {}
    
    def provision_device(self, device_type: str, location: str) -> Dict:
        """Provision new device"""
        device_id = str(uuid4())
        
        device = {
            'device_id': device_id,
            'type': device_type,
            'location': location,
            'status': 'active',
            'created_at': datetime.now().isoformat(),
            'last_updated': datetime.now().isoformat()
        }
        
        self.devices[device_id] = device
        
        # Generate device certificate
        cert = self._generate_certificate(device_id)
        self.certificates[device_id] = cert
        
        print(f"Provisioned device: {device_id}")
        
        return {
            'device_id': device_id,
            'certificate': cert['public_key'],
            'endpoint': 'mqtt://iot.example.com'
        }
    
    def _generate_certificate(self, device_id: str) -> Dict:
        """Generate device certificate"""
        # In production, use proper certificate generation
        return {
            'public_key': f'-----BEGIN CERTIFICATE-----\n{device_id}\n-----END CERTIFICATE-----',
            'private_key': f'-----BEGIN PRIVATE KEY-----\n{device_id}\n-----END PRIVATE KEY-----'
        }
    
    def revoke_device(self, device_id: str):
        """Revoke device"""
        if device_id in self.devices:
            self.devices[device_id]['status'] = 'revoked'
            print(f"Revoked device: {device_id}")

class DeviceManagementService:
    """Manage device firmware, configuration"""
    
    def __init__(self):
        self.firmware_versions = {}
        self.device_configs = {}
    
    def push_firmware_update(self, device_id: str, firmware_version: str):
        """Push firmware update to device"""
        update_payload = {
            'type': 'firmware_update',
            'version': firmware_version,
            'url': f'https://updates.example.com/{firmware_version}.bin',
            'checksum': self._calculate_checksum(firmware_version)
        }
        
        print(f"Pushing firmware update to {device_id}: v{firmware_version}")
        
        # In real system, would send via MQTT
        return update_payload
    
    def update_device_config(self, device_id: str, config: Dict):
        """Update device configuration"""
        self.device_configs[device_id] = config
        
        print(f"Updated config for {device_id}: {config}")
    
    def _calculate_checksum(self, version: str) -> str:
        """Calculate firmware checksum"""
        import hashlib
        return hashlib.md5(version.encode()).hexdigest()
```

### Device Health Monitoring
```python
class DeviceHealthMonitor:
    """Monitor health and status of IoT devices"""
    
    def __init__(self):
        self.health_data = {}
        self.alerts = []
    
    def record_health_metric(self, device_id: str, metric: str, value: float):
        """Record health metric from device"""
        if device_id not in self.health_data:
            self.health_data[device_id] = {}
        
        self.health_data[device_id][metric] = {
            'value': value,
            'timestamp': datetime.now().isoformat()
        }
        
        # Check for anomalies
        self._check_health_thresholds(device_id, metric, value)
    
    def _check_health_thresholds(self, device_id: str, metric: str, value: float):
        """Check if metric exceeds thresholds"""
        thresholds = {
            'temperature': {'max': 80, 'min': 0},
            'cpu_usage': {'max': 95, 'min': 0},
            'memory_usage': {'max': 95, 'min': 0},
            'battery_level': {'max': 100, 'min': 10}
        }
        
        if metric in thresholds:
            limits = thresholds[metric]
            
            if value > limits['max']:
                self._raise_alert(device_id, f"{metric} too high: {value}")
            elif value < limits['min']:
                self._raise_alert(device_id, f"{metric} too low: {value}")
    
    def _raise_alert(self, device_id: str, message: str):
        """Raise alert for device issue"""
        alert = {
            'device_id': device_id,
            'message': message,
            'timestamp': datetime.now().isoformat(),
            'severity': 'warning'
        }
        
        self.alerts.append(alert)
        print(f"⚠️  ALERT for {device_id}: {message}")
    
    def get_device_status(self, device_id: str) -> Dict:
        """Get overall device status"""
        if device_id not in self.health_data:
            return {'status': 'unknown'}
        
        metrics = self.health_data[device_id]
        
        # Determine overall health
        if self._has_critical_issues(device_id):
            status = 'critical'
        elif self._has_warnings(device_id):
            status = 'warning'
        else:
            status = 'healthy'
        
        return {'status': status, 'metrics': metrics}
    
    def _has_critical_issues(self, device_id: str) -> bool:
        """Check for critical issues"""
        return any(a['device_id'] == device_id for a in self.alerts)
    
    def _has_warnings(self, device_id: str) -> bool:
        """Check for warnings"""
        return len([a for a in self.alerts if a['device_id'] == device_id]) > 0
```
</VSCode.Cell>

<VSCode.Cell language="markdown">
## 3. Edge ML Inference

### On-Device Model Inference
```python
class EdgeMLInference:
    """Run ML models on edge devices"""
    
    def __init__(self, model_path: str, use_gpu: bool = False):
        self.model_path = model_path
        self.use_gpu = use_gpu
        self.model = None
        self.load_model()
    
    def load_model(self):
        """Load quantized model on edge device"""
        # Use quantized/compressed models for edge
        # e.g., TensorFlow Lite, ONNX Runtime
        print(f"Loading model from {self.model_path}")
        
        # Simulate model loading
        self.model = {'model_id': 'edge_model_v1'}
    
    def preprocess_input(self, raw_input):
        """Preprocess input for model"""
        # Resize image, normalize, etc.
        return {'processed': True}
    
    def inference(self, input_data) -> Dict:
        """Run inference on edge device"""
        import time
        
        # Preprocess
        processed = self.preprocess_input(input_data)
        
        # Run inference (typically <100ms on edge)
        start = time.time()
        # prediction = self.model.predict(processed)
        prediction = {'class': 'positive', 'confidence': 0.95}
        elapsed = time.time() - start
        
        return {
            'prediction': prediction,
            'latency_ms': elapsed * 1000,
            'model_version': 'v1'
        }
    
    def get_model_size(self) -> int:
        """Get model size in MB"""
        # Quantized models typically 2-50MB
        return 5  # MB

class EdgeMLOptimization:
    """Optimize models for edge deployment"""
    
    @staticmethod
    def quantize_model(model, bits: int = 8):
        """Quantize model to reduce size"""
        # Convert float32 to int8
        print(f"Quantizing model to {bits}-bit")
        return model
    
    @staticmethod
    def prune_model(model, pruning_ratio: float = 0.5):
        """Prune model to reduce size"""
        print(f"Pruning {pruning_ratio*100}% of model weights")
        return model
    
    @staticmethod
    def distill_model(teacher_model, student_size: str = 'small'):
        """Distill larger model to smaller student"""
        print(f"Distilling model to {student_size} size")
        return student_size
```

### Federated Learning on Edge
```python
class FederatedLearningCoordinator:
    """Coordinate federated learning across edge devices"""
    
    def __init__(self):
        self.global_model_version = 0
        self.device_updates = {}
        self.aggregation_strategy = 'fedavg'
    
    def start_training_round(self, devices: List[str]) -> Dict:
        """Start federated training round"""
        print(f"Starting federated learning round {self.global_model_version}")
        
        return {
            'round': self.global_model_version,
            'devices': devices,
            'model_version': self.global_model_version,
            'local_epochs': 5
        }
    
    def receive_device_update(self, device_id: str, update: Dict):
        """Receive model update from device"""
        self.device_updates[device_id] = {
            'model_update': update,
            'timestamp': datetime.now().isoformat()
        }
        
        print(f"Received update from {device_id}")
    
    def aggregate_updates(self, participating_devices: List[str]) -> Dict:
        """Aggregate updates using FedAvg"""
        updates = [
            self.device_updates[d]['model_update'] 
            for d in participating_devices 
            if d in self.device_updates
        ]
        
        # Average the updates
        aggregated = self._fedavg_aggregate(updates)
        
        self.global_model_version += 1
        
        return aggregated
    
    def _fedavg_aggregate(self, updates: List[Dict]) -> Dict:
        """Federated Averaging aggregation"""
        # Simplified: in production, would handle tensors properly
        return {'aggregated': True, 'version': self.global_model_version}
```
</VSCode.Cell>

<VSCode.Cell language="markdown">
## 4. Offline-First Synchronization

### Offline Data Sync
```python
from collections import deque

class OfflineFirstSync:
    """Sync data with cloud while supporting offline operation"""
    
    def __init__(self):
        self.local_storage = {}
        self.sync_queue = deque()
        self.is_online = True
    
    def write_data(self, key: str, value: any):
        """Write data (works offline)"""
        self.local_storage[key] = {
            'value': value,
            'timestamp': datetime.now().isoformat(),
            'synced': False
        }
        
        # Queue for sync
        self.sync_queue.append({
            'operation': 'write',
            'key': key,
            'value': value,
            'timestamp': datetime.now().isoformat()
        })
    
    def read_data(self, key: str):
        """Read data (returns local copy if offline)"""
        if key in self.local_storage:
            return self.local_storage[key]['value']
        return None
    
    def sync_with_cloud(self) -> int:
        """Sync pending changes with cloud"""
        if not self.is_online:
            print("Not connected to cloud. Queuing updates.")
            return 0
        
        synced_count = 0
        
        while self.sync_queue:
            operation = self.sync_queue.popleft()
            
            try:
                # Send to cloud
                self._send_to_cloud(operation)
                
                # Mark as synced
                key = operation['key']
                if key in self.local_storage:
                    self.local_storage[key]['synced'] = True
                
                synced_count += 1
                
            except Exception as e:
                # Put back in queue on failure
                self.sync_queue.appendleft(operation)
                print(f"Sync failed: {str(e)}")
                break
        
        return synced_count
    
    def _send_to_cloud(self, operation: Dict):
        """Send operation to cloud"""
        print(f"Syncing: {operation['operation']} {operation['key']}")
    
    def set_connectivity(self, is_online: bool):
        """Update connectivity status"""
        self.is_online = is_online
        
        if is_online:
            print("Online. Syncing...")
            self.sync_with_cloud()
        else:
            print("Offline. Queueing updates...")
```
</VSCode.Cell>

<VSCode.Cell language="markdown">
## 5. Edge Computing Checklist

✅ **Architecture**
- [ ] Edge deployment model defined (fog/edge cloud)
- [ ] Latency requirements met (<50ms typical)
- [ ] Edge-cloud coordination working
- [ ] Failover to cloud implemented
- [ ] Scalability verified

✅ **Device Management**
- [ ] Device provisioning automated
- [ ] Firmware update mechanism working
- [ ] Configuration management centralized
- [ ] Device health monitoring enabled
- [ ] Remote troubleshooting possible

✅ **Edge ML**
- [ ] Models quantized for edge
- [ ] Inference latency <100ms
- [ ] Federated learning pipeline working
- [ ] Model versioning managed
- [ ] Local inference fallback available

✅ **Synchronization**
- [ ] Offline-first sync implemented
- [ ] Conflict resolution working
- [ ] Eventual consistency achieved
- [ ] Bandwidth usage optimized
- [ ] Data integrity maintained

✅ **Security**
- [ ] Device certificates provisioned
- [ ] Communication encrypted
- [ ] Data at rest encrypted
- [ ] Access control enforced
- [ ] Audit logs maintained
</VSCode.Cell>
```
````
