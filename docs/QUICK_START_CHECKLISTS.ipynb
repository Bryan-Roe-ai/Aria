# Aria Quick-Start Checklists

**Role-specific, step-by-step checklists to get productive quickly.**

Use these checklists to onboard quickly or set up a new development environment. Each checklist is designed to be completed in 30-60 minutes.

## 👤 New Developer - First Day Checklist

**Goal:** Understand the codebase structure and run your first test.
**Time:** 45 minutes
**Outcome:** Can navigate the codebase and run tests

### Phase 1: Setup (15 min)
- [ ] Clone repository: `git clone https://github.com/Bryan-Roe/Aria.git`
- [ ] Navigate to repo: `cd /workspaces/Aria`
- [ ] Verify Python 3.12: `python --version`
- [ ] Create virtual environment: `python -m venv .venv`
- [ ] Activate venv: `source .venv/bin/activate`
- [ ] Install dependencies: `pip install -r requirements.txt`
- [ ] Verify installation: `python -c "import torch, peft, qiskit; print('✓ All deps installed')"`

### Phase 2: Documentation (15 min)
- [ ] Read: `docs/REPOSITORY_TOPICS.md` (skim for overview)
- [ ] Read: `docs/MASTER_INDEX.md` (understand structure)
- [ ] Bookmark: All docs in `docs/` folder
- [ ] Note: Key files to watch (function_app.py, shared/, apps/)

### Phase 3: First Run (15 min)
- [ ] Check health: `curl http://localhost:7071/api/ai/status | jq` (if Functions running)
- [ ] Run unit tests: `python scripts/test_runner.py --unit`
- [ ] View results: Should see ~2690 tests pass, ~6 expected failures
- [ ] Check coverage: `python scripts/test_runner.py --unit --coverage`

### Success Metrics
- ✓ Can run `python scripts/test_runner.py --unit` without errors
- ✓ Can access docs and search them with Ctrl+F
- ✓ Can explain 3 major components of Aria
- ✓ Identified which project you'll work on

## 💻 Feature Developer - First Feature Checklist

**Goal:** Understand where to add a new feature and have a clear plan.
**Time:** 60 minutes
**Outcome:** Clear implementation plan with files to modify

### Phase 1: Feature Understanding (15 min)
- [ ] Read the feature description/issue
- [ ] Identify the category: Chat? Character? Training? Quantum?
- [ ] Search `docs/REPOSITORY_TOPICS.md` for that category
- [ ] List 3 key files that might need changes

### Phase 2: Dependency Analysis (15 min)
- [ ] Open `docs/REPOSITORY_ARCHITECTURE.md`
- [ ] Find your feature area in the dependency graph
- [ ] List all dependencies (what does my feature depend on?)
- [ ] List all dependents (what depends on my feature?)
- [ ] Check for conflicts or circular dependencies

### Phase 3: File Location Mapping (15 min)
- [ ] Use `docs/KEYWORD_INDEX.md` to find exact file locations
- [ ] Search: Main feature keyword (e.g., "memory", "quantum", "provider")
- [ ] List exact files to modify
- [ ] List test files that test those modules
- [ ] Note: Related configuration files

### Phase 4: Architecture Review (15 min)
- [ ] Find relevant diagram in `docs/ARCHITECTURE_DIAGRAMS.md`
- [ ] Understand the process flow
- [ ] Mark where your feature fits in the flow
- [ ] Note: Input/output contracts
- [ ] Check: Error handling requirements

### Implementation Plan Ready
- [ ] Can list 5+ files to modify (ordered by dependency)
- [ ] Can draw the modified process flow
- [ ] Can identify 3+ test cases needed
- [ ] Can list 1-2 potential issues/edge cases
- [ ] Have clear success criteria

## 🏗️ System Architect - Design Review Checklist

**Goal:** Validate new architecture before implementation starts.
**Time:** 90 minutes (can be split into sessions)
**Outcome:** Approved design with clear rationale

### Phase 1: Problem Understanding (20 min)
- [ ] Read design document/proposal
- [ ] Identify problem statement (what are we solving?)
- [ ] List constraints (performance, security, cost, etc.)
- [ ] Note: Non-functional requirements (scalability, reliability, etc.)

### Phase 2: Architecture Analysis (20 min)
- [ ] Review: REPOSITORY_ARCHITECTURE.md (full document)
- [ ] Check: Proposed design against existing architecture
- [ ] Identify: Components to modify vs. new components
- [ ] Note: Any architecture violations or deviations

### Phase 3: Dependency Review (20 min)
- [ ] Map: New component dependencies
- [ ] Check: Against existing dependency graph
- [ ] Verify: No circular dependencies
- [ ] Validate: Interfaces between components
- [ ] Confirm: No breaking changes to existing APIs

### Phase 4: Risk Assessment (20 min)
- [ ] Identify: High-risk areas (quantum, database, auth, etc.)
- [ ] List: Failure modes and mitigation strategies
- [ ] Check: Error handling paths
- [ ] Confirm: Monitoring/observability in place
- [ ] Note: Rollback procedures

### Phase 5: Pattern Consistency (10 min)
- [ ] Verify: Follows established patterns (see Section 4 of Interactive Guide)
- [ ] Check: Naming conventions consistent
- [ ] Confirm: Error handling matches existing code
- [ ] Verify: Logging/telemetry included

### Design Review Complete
- [ ] Clear problem statement documented
- [ ] All dependencies mapped
- [ ] Risk assessment completed
- [ ] Pattern consistency verified
- [ ] Ready to assign to implementation team

## 🐛 Debugger - Debugging Workflow Checklist

**Goal:** Systematically diagnose and fix a bug.
**Time:** 30-60 minutes (depends on complexity)
**Outcome:** Root cause identified and fix verified

### Phase 1: Problem Triage (10 min)
- [ ] Reproduce the issue (can I make it happen consistently?)
- [ ] Note: Error message, stack trace, or observable symptoms
- [ ] Identify: Affected component(s) (chat? quantum? training?)
- [ ] Check: When did it start? (recent change? regression?)

### Phase 2: Quick Diagnosis (10 min)
- [ ] Check: `/api/ai/status` health check
- [ ] Verify: Providers are working
- [ ] Check: Database connection
- [ ] Verify: Required env vars are set
- [ ] Note: Any obvious configuration issues

### Phase 3: Deep Diagnosis (15 min)
- [ ] Use: `docs/REPOSITORY_ARCHITECTURE.md` Part 5 (Problem Finder)
- [ ] Search problem description (exact or similar)
- [ ] Read: Suggested solutions
- [ ] Find: File(s) to examine
- [ ] Get: Additional context from code comments

### Phase 4: Investigation (15 min)
- [ ] Open the identified files
- [ ] Add logging/breakpoints as needed
- [ ] Run: Minimal test case to reproduce
- [ ] Trace: Execution path
- [ ] Identify: Exact line causing the issue

### Phase 5: Fix & Verify (10 min)
- [ ] Make: Minimal fix (don't refactor)
- [ ] Run: Unit test for that module
- [ ] Run: Related integration tests
- [ ] Verify: Issue is resolved
- [ ] Note: Root cause for documentation

### Bug Fix Complete
- [ ] Issue reproduced initially
- [ ] Root cause identified
- [ ] Fix verified with tests
- [ ] No regressions introduced
- [ ] Documentation updated if needed

## ⚛️ Quantum ML Developer - Setup Checklist

**Goal:** Get quantum development environment working locally.
**Time:** 45 minutes
**Outcome:** Can run quantum circuits and see results

### Prerequisites
- [ ] Python 3.12 installed
- [ ] Virtual environment created and activated
- [ ] Requirements installed: `pip install -r requirements.txt`

### Phase 1: Quantum Dependencies (10 min)
- [ ] Verify Qiskit: `python -c "import qiskit; print(qiskit.__version__)"`
- [ ] Verify PEFT: `python -c "import peft; print('PEFT OK')"`
- [ ] Verify Pennylane (optional): `python -c "import pennylane; print('PL OK')"` (if using)
- [ ] Note: Versions in console

### Phase 2: Quantum Configuration (15 min)
- [ ] Set simulator-only mode: `export QUANTUM_PROVIDER="simulator"`
- [ ] Disable cost checks: `export AZURE_CONFIRM_COST="false"`
- [ ] Set max shots: `export QUANTUM_MAX_SHOTS="100"`
- [ ] Create config file: `config/quantum_dev.yaml` (copy from `quantum_autorun.yaml`)
- [ ] Edit: Set `simulator_only: true`

### Phase 3: First Quantum Run (10 min)
- [ ] Run dry-run: `python scripts/quantum_autorun.py --dry-run`
- [ ] Verify: Config loads without errors
- [ ] Run test: `pytest tests/unit/test_quantum_* -v`
- [ ] Check: All tests pass (or note expected failures)

### Phase 4: MCP Server (10 min)
- [ ] Start MCP server: `python ai-projects/quantum-ml/quantum_mcp_server.py`
- [ ] Verify: Server starts and listens
- [ ] Note: Server port and health endpoint
- [ ] Test: Simple API call

### Quantum Setup Complete
- [ ] Can run quantum tests locally
- [ ] MCP server runs without errors
- [ ] Understand simulator vs. real QPU trade-offs
- [ ] Know cost control settings

## 🎓 Training Specialist - Autonomous Training Setup

**Goal:** Get autonomous training orchestrator running and understand its lifecycle.
**Time:** 30 minutes
**Outcome:** Training cycles running and can monitor progress

### Phase 1: Environment Setup (10 min)
- [ ] Check datasets exist: `ls datasets/quantum datasets/chat datasets/massive_quantum`
- [ ] Verify base model: `ls ai-projects/lora-training/microsoft_phi-silica-3.6_v1/`
- [ ] Check output folder: `mkdir -p data_out/autonomous_training`
- [ ] Verify config: `cat config/autonomous_training.yaml | head -20`

### Phase 2: Configuration Review (10 min)
- [ ] Open: `config/autonomous_training.yaml`
- [ ] Check: `cycle_interval_minutes` (default: 30)
- [ ] Check: `epochs_progression` (default: [25, 50, 100, 200])
- [ ] Check: `auto_promote` (enable if you want auto-deployment)
- [ ] Note: Any custom settings needed

### Phase 3: Start Training (10 min)
- [ ] Start orchestrator: `python scripts/autonomous_training_orchestrator.py`
- [ ] Verify: Logs show "Cycle started" message
- [ ] Monitor: `tail -f data_out/autonomous_training.log`
- [ ] Check: Status file: `cat data_out/autonomous_training_status.json | jq`
- [ ] Note: Cycle start time and expected duration

### Training Running
- [ ] Cycles running successfully
- [ ] Logs show progress
- [ ] Status JSON updating
- [ ] Can force cycle with: `pkill -USR1 -f autonomous_training_orchestrator.py`

## 📋 Common Checklist Patterns

### Before Every Commit
- [ ] Run tests: `python scripts/test_runner.py --unit`
- [ ] Check linting: `make lint` (or `ruff check .`)
- [ ] Verify no debug statements left in code
- [ ] Update CHANGELOG.md if user-facing change
- [ ] Verify: No secrets committed (check .env files)

### Before Merging PR
- [ ] All tests passing
- [ ] Coverage maintained or improved (>80%)
- [ ] Code reviewed by 1+ team member
- [ ] Documentation updated (code comments + docs/)
- [ ] No breaking changes to APIs

### Before Deploying to Production
- [ ] Full test suite passes: `python scripts/test_runner.py --all`
- [ ] Performance baseline met (check PERFORMANCE_OPTIMIZATION_GUIDE.md)
- [ ] Database migrations tested (if any)
- [ ] Rollback procedure documented
- [ ] Monitoring alerts configured
- [ ] Stakeholders notified

### Before Adding New Dependencies
- [ ] Check for security vulnerabilities
- [ ] Verify compatibility with existing deps
- [ ] Check file size impact
- [ ] Document reason for new dependency
- [ ] Update requirements.txt and requirements-dev.txt
