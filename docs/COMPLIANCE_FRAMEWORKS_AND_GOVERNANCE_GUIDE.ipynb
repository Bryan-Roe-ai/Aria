````python
```xml
<VSCode.Cell language="markdown">
# ✅ Compliance Frameworks & Governance Guide

**Implementing SOC 2, ISO 27001, HIPAA, GDPR, and other compliance frameworks**

This guide covers:
- SOC 2 Type I & II audits
- ISO 27001 information security management
- HIPAA healthcare data compliance
- PCI DSS payment card compliance
- GDPR data protection (EU)
- CCPA privacy (California)
- Audit trails and logging
- Compliance automation and tooling
</VSCode.Cell>

<VSCode.Cell language="markdown">
## 1. SOC 2 Compliance

### Security Controls Implementation
```python
from datetime import datetime, timedelta
import logging

class SOC2Compliance:
    """SOC 2 Type II control implementation"""
    
    def __init__(self):
        self.audit_log = []
        self.access_controls = {}
    
    def log_access(self, user_id: str, resource: str, action: str, status: str):
        """CC6.1: Logical access control"""
        log_entry = {
            'timestamp': datetime.utcnow(),
            'user_id': user_id,
            'resource': resource,
            'action': action,
            'status': status,
            'ip_address': self._get_client_ip()
        }
        
        self.audit_log.append(log_entry)
        logging.info(f"Access: {user_id} -> {resource} ({status})")
    
    def enforce_mfa(self, user_id: str) -> bool:
        """A1.2: MFA enforcement"""
        # Check if MFA is enabled for user
        mfa_status = self.access_controls.get(user_id, {}).get('mfa_enabled', False)
        
        if not mfa_status:
            logging.warning(f"MFA not enabled for {user_id}")
            return False
        
        return True
    
    def encrypt_data_at_rest(self, data: str, key: bytes) -> bytes:
        """P8.1: Data encryption"""
        from cryptography.fernet import Fernet
        
        cipher = Fernet(key)
        encrypted = cipher.encrypt(data.encode())
        
        return encrypted
    
    def encrypt_data_in_transit(self, connection):
        """C1.2: Communication encryption (TLS)"""
        # Enforce TLS 1.2 or higher
        assert connection.tls_version >= 1.2
        return True
    
    def change_management_log(self, change_id: str, description: str, approver: str):
        """C2.1: Change management"""
        change = {
            'change_id': change_id,
            'description': description,
            'approver': approver,
            'timestamp': datetime.utcnow(),
            'status': 'approved'
        }
        
        logging.info(f"Change {change_id} approved by {approver}")
        return change
    
    def incident_response_procedure(self, incident_id: str, severity: str):
        """C4.2: Incident response"""
        incident = {
            'incident_id': incident_id,
            'severity': severity,
            'discovered_at': datetime.utcnow(),
            'status': 'investigating',
            'timeline': []
        }
        
        logging.warning(f"Incident {incident_id} ({severity}) initiated")
        return incident
    
    def _get_client_ip(self):
        """Get client IP for audit"""
        import socket
        return socket.gethostbyname(socket.gethostname())

# Usage
soc2 = SOC2Compliance()
soc2.log_access('user@example.com', '/api/users', 'GET', 'success')
soc2.enforce_mfa('user@example.com')
```

### SOC 2 Readiness Checklist
```python
class SOC2ReadinessChecker:
    """Assess SOC 2 compliance readiness"""
    
    CONTROLS = {
        'CC1': 'General Deterrents',
        'CC2': 'Change Management',
        'CC3': 'Separation of Duties',
        'CC4': 'Restricted Access',
        'CC5': 'Access Credentials',
        'CC6': 'Logical Access Controls',
        'CC7': 'System Monitoring',
        'CC8': 'Recovery Procedures',
        'CC9': 'Risk Mitigation'
    }
    
    def check_readiness(self):
        """Generate readiness report"""
        readiness = {}
        
        for control, description in self.CONTROLS.items():
            readiness[control] = {
                'description': description,
                'implemented': False,
                'evidence': [],
                'remediation': ''
            }
        
        return readiness
```
</VSCode.Cell>

<VSCode.Cell language="markdown">
## 2. ISO 27001 Implementation

### Information Security Management System
```python
class ISO27001ISMS:
    """ISO 27001 Information Security Management System"""
    
    def __init__(self):
        self.policies = {}
        self.assets = {}
        self.risks = {}
        self.controls = {}
    
    def register_asset(self, asset_id: str, asset_type: str, classification: str):
        """A.8.1.1: Asset inventory"""
        asset = {
            'asset_id': asset_id,
            'type': asset_type,
            'classification': classification,  # Public, Internal, Confidential, Secret
            'registered_date': datetime.utcnow(),
            'owner': None
        }
        
        self.assets[asset_id] = asset
        return asset
    
    def conduct_risk_assessment(self, asset_id: str, threat: str, probability: int, impact: int):
        """A.12.6.1: Risk assessment and treatment"""
        risk_score = probability * impact
        
        risk = {
            'asset_id': asset_id,
            'threat': threat,
            'probability': probability,
            'impact': impact,
            'risk_score': risk_score,
            'status': 'assessed'
        }
        
        self.risks[asset_id] = risk
        return risk
    
    def implement_control(self, control_id: str, control_name: str, objective: str):
        """A.13: Implement security controls"""
        control = {
            'control_id': control_id,
            'name': control_name,
            'objective': objective,
            'implementation_date': datetime.utcnow(),
            'effectiveness': None
        }
        
        self.controls[control_id] = control
        return control
    
    def audit_and_review(self, audit_id: str, scope: str):
        """Clause 9.2: Internal audits"""
        audit = {
            'audit_id': audit_id,
            'scope': scope,
            'start_date': datetime.utcnow(),
            'findings': [],
            'status': 'in_progress'
        }
        
        return audit

# Usage
iso27001 = ISO27001ISMS()
iso27001.register_asset('db-prod-1', 'database', 'Confidential')
iso27001.conduct_risk_assessment('db-prod-1', 'SQL Injection', 3, 4)
iso27001.implement_control('C1', 'Input Validation', 'Prevent injection attacks')
```

### Security Policy Template
```python
class SecurityPolicy:
    """Template for organizational security policies"""
    
    POLICY_TYPES = [
        'Access Control Policy',
        'Incident Management Policy',
        'Change Management Policy',
        'Data Classification Policy',
        'Backup and Recovery Policy',
        'Third-party Management Policy'
    ]
    
    def create_policy(self, policy_type: str, version: str):
        """Create security policy document"""
        policy = {
            'type': policy_type,
            'version': version,
            'created_date': datetime.utcnow(),
            'approved_by': None,
            'effective_date': None,
            'review_date': None,
            'content': self._generate_policy_content(policy_type)
        }
        
        return policy
    
    def _generate_policy_content(self, policy_type: str) -> dict:
        """Generate policy content template"""
        return {
            'scope': '',
            'purpose': '',
            'policy_statement': '',
            'procedures': [],
            'responsibilities': {},
            'enforcement': '',
            'review_frequency': 'Annually'
        }
```
</VSCode.Cell>

<VSCode.Cell language="markdown">
## 3. GDPR & Data Privacy

### Data Subject Rights Implementation
```python
class GDPRCompliance:
    """GDPR Article compliance implementation"""
    
    def __init__(self, data_store):
        self.data_store = data_store
    
    def right_to_access(self, subject_id: str) -> dict:
        """Article 15: Right of access"""
        personal_data = self.data_store.get_subject_data(subject_id)
        
        access_record = {
            'subject_id': subject_id,
            'request_date': datetime.utcnow(),
            'data_provided': personal_data,
            'provided_date': datetime.utcnow()
        }
        
        return access_record
    
    def right_to_erasure(self, subject_id: str) -> bool:
        """Article 17: Right to be forgotten"""
        # Check if erasure is legal
        if self._can_erase(subject_id):
            self.data_store.delete_subject_data(subject_id)
            logging.info(f"Data for {subject_id} erased per GDPR Article 17")
            return True
        
        return False
    
    def data_portability(self, subject_id: str, format: str = 'json') -> str:
        """Article 20: Data portability"""
        data = self.data_store.get_subject_data(subject_id)
        
        if format == 'json':
            return json.dumps(data)
        elif format == 'csv':
            return self._to_csv(data)
        
        return None
    
    def process_consent(self, subject_id: str, processing_purpose: str, consent: bool):
        """Article 6: Lawful basis - Consent"""
        consent_record = {
            'subject_id': subject_id,
            'purpose': processing_purpose,
            'consent': consent,
            'timestamp': datetime.utcnow(),
            'ip_address': self._get_client_ip()
        }
        
        logging.info(f"Consent recorded for {subject_id}: {consent}")
        return consent_record
    
    def data_breach_notification(self, breach_id: str, affected_subjects: int):
        """Article 33: Breach notification (72 hour requirement)"""
        deadline = datetime.utcnow() + timedelta(hours=72)
        
        notification = {
            'breach_id': breach_id,
            'discovery_date': datetime.utcnow(),
            'notification_deadline': deadline,
            'affected_subjects': affected_subjects,
            'status': 'notifying'
        }
        
        logging.warning(f"Breach {breach_id} - notify {affected_subjects} subjects by {deadline}")
        return notification
    
    def _can_erase(self, subject_id: str) -> bool:
        """Check legal basis for erasure"""
        # Example: check if retention period expired
        return True
    
    def _get_client_ip(self):
        import socket
        return socket.gethostbyname(socket.gethostname())
    
    def _to_csv(self, data: dict) -> str:
        """Convert to CSV format"""
        import csv, io
        output = io.StringIO()
        writer = csv.DictWriter(output, fieldnames=data.keys())
        writer.writeheader()
        writer.writerow(data)
        return output.getvalue()
```

### Data Processing Agreement
```python
class DataProcessingAgreement:
    """DPA for processors and controllers"""
    
    def __init__(self, controller_name: str, processor_name: str):
        self.controller = controller_name
        self.processor = processor_name
        self.terms = {}
    
    def set_processing_scope(self, scope: dict):
        """Define what data is processed"""
        self.terms['scope'] = {
            'subject_matter': scope.get('subject_matter'),
            'duration': scope.get('duration'),
            'nature': scope.get('nature'),
            'purposes': scope.get('purposes'),
            'data_categories': scope.get('data_categories'),
            'special_categories': scope.get('special_categories', False)
        }
    
    def set_security_measures(self, measures: list):
        """Article 32: Technical and organizational measures"""
        self.terms['security'] = {
            'encryption': 'AES-256',
            'access_controls': measures,
            'audit_logs': True,
            'employee_training': True,
            'incident_response': True
        }
```
</MySCode.Cell>

<MySCode.Cell language="markdown">
## 4. Compliance Automation

### Automated Compliance Monitoring
```python
class ComplianceMonitoring:
    """Continuous compliance monitoring"""
    
    def __init__(self):
        self.checks = []
        self.violations = []
    
    def add_compliance_check(self, check_name: str, check_func, severity: str = 'high'):
        """Register compliance check"""
        self.checks.append({
            'name': check_name,
            'func': check_func,
            'severity': severity,
            'last_run': None,
            'status': 'not_run'
        })
    
    def run_all_checks(self) -> dict:
        """Run all compliance checks"""
        results = {}
        
        for check in self.checks:
            try:
                result = check['func']()
                check['last_run'] = datetime.utcnow()
                check['status'] = 'passed' if result else 'failed'
                results[check['name']] = result
                
                if not result:
                    self.violations.append({
                        'check': check['name'],
                        'timestamp': datetime.utcnow(),
                        'severity': check['severity']
                    })
            
            except Exception as e:
                logging.error(f"Check {check['name']} failed: {str(e)}")
                check['status'] = 'error'
        
        return results
    
    def generate_compliance_report(self) -> dict:
        """Generate compliance status report"""
        total_checks = len(self.checks)
        passed = sum(1 for c in self.checks if c['status'] == 'passed')
        
        return {
            'timestamp': datetime.utcnow(),
            'total_checks': total_checks,
            'passed': passed,
            'failed': total_checks - passed,
            'compliance_percentage': (passed / total_checks * 100) if total_checks > 0 else 0,
            'violations': self.violations[-10:]  # Last 10 violations
        }
```
</MySCode.Cell>

<MySCode.Cell language="markdown">
## 5. Compliance Checklist

✅ **Governance**
- [ ] Compliance roadmap created
- [ ] Compliance team assigned
- [ ] Policies documented
- [ ] Risk assessment completed
- [ ] Controls mapped to requirements

✅ **Implementation**
- [ ] Access controls in place
- [ ] Encryption enabled
- [ ] Audit logging active
- [ ] Incident response plan ready
- [ ] Data classification done

✅ **Monitoring**
- [ ] Continuous monitoring active
- [ ] Compliance checks automated
- [ ] Dashboards configured
- [ ] Alerts configured
- [ ] Reports generated regularly

✅ **Audit**
- [ ] Internal audit scheduled
- [ ] Evidence collected
- [ ] Findings documented
- [ ] Remediation tracked
- [ ] Audit communication plan

✅ **Certification**
- [ ] Assessment scheduled
- [ ] Auditor selected
- [ ] Evidence package prepared
- [ ] Management sign-off obtained
- [ ] Certification maintained
</MySCode.Cell>
```
````
