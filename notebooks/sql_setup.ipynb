# SQL Setup for AI

This notebook demonstrates how to connect to the local SQLite database used by Aria.


```python
import sqlite3
from pathlib import Path

# Path from shared.sql_engine logic
sql_path = Path("data_out/qai_local.db")
print(f"Connecting to {sql_path}")
conn = sqlite3.connect(sql_path)
cur = conn.cursor()
cur.execute("SELECT name FROM sqlite_master WHERE type='table' AND name='sql_setup_check'")
print(cur.fetchone())
cur.execute("SELECT COUNT(*) FROM sql_setup_check")
print(f"Rows in sql_setup_check: {cur.fetchone()[0]}")
```

The table `sql_setup_check` should contain a row inserted during setup.
