# SQL Setup for AI

This notebook demonstrates how to connect to the local SQLite
database used by Aria.


```python
import sqlite3
from pathlib import Path

# Find database path relative to cwd or one of its parents.
candidate_roots = [Path.cwd(), *Path.cwd().parents]
sql_path = None
for root in candidate_roots:
    candidate = root / "data_out" / "qai_local.db"
    if candidate.exists():
        sql_path = candidate
        break

if sql_path is None:
    sql_path = Path.cwd() / "data_out" / "qai_local.db"
    sql_path.parent.mkdir(parents=True, exist_ok=True)

print(f"Connecting to {sql_path.resolve()}")
conn = sqlite3.connect(sql_path)
cur = conn.cursor()
cur.execute("SELECT name FROM sqlite_master WHERE type='table' AND name='sql_setup_check'")
print(cur.fetchone())
cur.execute("SELECT COUNT(*) FROM sql_setup_check")
print(f"Rows in sql_setup_check: {cur.fetchone()[0]}")
```

The table `sql_setup_check` should contain a row inserted during setup.
